```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
data_url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
boston_df = pd.read_csv(data_url)
```

```python
sns.set(rc={'figure.figsize':(10, 10)})
sns.displot(boston_df['medv'])
plt.show()
```

```python
correlation_matrix = boston_df.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
plt.figure(figsize=(20, 5))
features = ['lstat', 'rm']
target = boston_df['medv']
for i, col in enumerate(features):
    plt.subplot(1, len(features), i+1)
    x = boston_df[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('medv')
```

```python
X = pd.DataFrame(np.c_[boston_df['lstat'], boston_df['rm']], columns=['lstat', 'rm'])
Y = boston_df['medv']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=5)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, Y_train)
reg.predict(X_test)
print('R2: ',  reg.score(X_test, Y_test))
```

```python
Y_pred = reg.predict(X_test)
plt.scatter(Y_pred, Y_test)
plt.xlabel('Y_pred')
plt.ylabel('Y_test')
plt.show()
```

```python
reg.intercept_
```

```python
coeff_df = pd.DataFrame(reg.coef_, X_train.columns, columns=['Coefficient'])
print(coeff_df)
```
